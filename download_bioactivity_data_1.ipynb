```python
# Import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

#  **Search for Target protein**


 **Alzheimer Related Target Search**

```python
# Target search for Alzheimer ((Beta-secretase 1, CHEMBL4822)
target = new_client.target
target_query = target.search('CHEMBL4822')
targets = pd.DataFrame.from_dict(target_query)
targets
```

**Select and retrieve bioactivity data for Beta-secretase 1 (first entry)**

```python
selected_target = targets.target_chembl_id[0]
selected_target
```

 Retrieve only bioactivity data for Beta-secretase 1 (CHEMBL4822) that are reported as IC50  values.

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df.head()
```

```python
# IC50 values are reported in nano molalar units within "standard_value" column
```

```python
#retrieves all the unique values in the standard_type column of your Pandas DataFrame (df
df.standard_type.unique()
```

```python
#save the resulting bioactivity data to a CSV file bioactivity_data.csv.
df.to_csv('beta_secretase1_bioactivity_data.csv', index=False)
```
